```
data = read.csv("台灣電力公司_太陽光電發電量及平均單位裝置容量每日發電量統計表.csv")
data
```

```
data_1 = subset(data,光電站名稱.site.name=="中大光電/Zhongda Photovoltaics") #核三廠光電
data_1 = subset(data_1,年度.year=="2016") #核三廠光電
data_1 = head(data_1,8)
data_2 = subset(data,光電站名稱.site.name=="龍井光電/Longjing Photovoltaics") #龍井光電
data_2 = subset(data_2,年度.year=="2016") #龍井光電
data_2 = head(data_2,8)
data_3 = subset(data,光電站名稱.site.name=="后里光電/Houli Photovoltaics") #后里光電
data_3 = subset(data_3,年度.year=="2016") #后里光電
data_3 = head(data_3,8)
data_4 = subset(data,光電站名稱.site.name=="金門光電/Kinme Photovoltaics") #金門光電
data_4 = subset(data_4,年度.year=="2016") #金門光電
data_4 = head(data_4,8)
```

```
data_1 %>% ggplot(aes(x = c("1","2","3","4","5","6","7","8"),y = 發電量.度..Power.generation.kW.h.	 ,fill= 月份.month	)) + geom_col() + geom_text(aes(label = 發電量.度..Power.generation.kW.h.	),vjust = -0.25) + labs(title = "中大光電不同月份發電量比較",x = "月份",y = "發電量") + theme(legend.position = "none")
```
