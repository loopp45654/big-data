```
install.packages("ggplot2")
install.packages("SportsAnalytics")
```

```
library(SportsAnalytics)
library(ggplot2)
```

```
NBA1819<-fetch_NBAPlayerStatistics("18-19")
str(NBA1819)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1819,color=Team)
qplot(Assists,Steals,data=NBA1819,color=Team)
```

```
qplot(TotalPoints,data=NBA1819,fill=Team)
qplot(Assists,data=NBA1819,fill=Team)
```

```
qplot(TotalPoints,data=NBA1819,geom="density",fill=Team)
qplot(Assists,data=NBA1819,geom="density",fill=Team)
```

```
qplot(TotalPoints,data=NBA1819,geom="boxplot",fill=Team)
qplot(Assists,data=NBA1819,geom="boxplot",fill=Team)
```
