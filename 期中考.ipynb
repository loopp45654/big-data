```
install.packages("SportsAnalytics")
library(SportsAnalytics)
NBA1617<-fetch_NBAPlayerStatistics("16-17")
```

```
head(NBA1617)
```

```
str(NBA1617)
```

```
NBA1617OrderM<-NBA1617[order(NBA1617$ThreesMade,decreasing=T),]
NBA1617OrderM[1:20,"ThreesMade"] ##取前二十名投三分球最多的數字並列出來
NBA1617OrderM[1:20,"Name"]    ##取前二十名投三分球最多的球員並列出來
```

```
subset(NBA1617,Team=="MEM") ##把熱火隊的球員列出來
```

```
NBA1617[grepl("James",NBA1617$Name),] ##把名字有James的人列出來
```

```
install.packages("ggplot2")
library(ggplot2)
```

```
##用點狀圖來表示各個球員的當季的進球數和當季的出手次數的情況,越左上方表示投籃越不准,越右下表示投籃越不准
ggplot(NBA1617,aes(x=FieldGoalsMade,y=FieldGoalsAttempted))+geom_point()
```

```
##這張圖表示球員的得分次數,而上方表示球員人數
ggplot(NBA1617,aes(x=FieldGoalsMade))+geom_bar()
```

```
##由圖形可知大部分的球員的進球數為100-300之間
ggplot(NBA1617,aes(x=FieldGoalsMade))+geom_boxplot()
```

```
##圖形表示各個球員的當季的進球數和當季的出手次數的情況,越左上方表示投籃越不准,越右下表示投籃越不准,且點越大的人數越多
ggplot(NBA1617,aes(x=FieldGoalsMade,y=FieldGoalsAttempted))+geom_count()
```
